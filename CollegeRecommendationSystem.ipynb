now we do text vectorization using BOW..

```python
# here we applly stemming first
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# define a function to perform stemming on the tags column

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
stem(collegeData['tags2'][0])
```

```python
collegeData['tags3']=collegeData['tags2'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(binary=True)
```

```python
Vectors=cv.fit_transform(collegeData['tags3']).toarray()
```

```python
len(Vectors)
```

```python
# for this we use cosine similarity.....

from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(Vectors).shape
```

```python
similarity=cosine_similarity(Vectors)
```

```python
similarity.shape
```

```python
similarity[0]   # it give the similarity of 1 college with all other colleges......similarity value lie between 0 to 1 so similarity with its own movie is 1
```

```python
# here we sort the similarity when a particular college is given to us ....but a problem occur thatit looses it index position so we us enumerate function  to get index position along with the similarity then we don't losse the index

sorted(similarity[0])[-10:-1]  #here it looses it index position her it is difficult to find that which college has this similarity
```

```python
# to solve this we use.....
sorted(list(enumerate(similarity[78])),reverse=True)  #it does sorting on the basis of index position  but we want it on similarity so we do
```

```python
sorted(list(enumerate(similarity[78])),reverse=True, key=lambda x:x[1])[1:11]   #here we get top 6 sorted list of movies based on similarity
```

```python
collegeData.columns
```

```python
# defining function for recommendation

def recommend(college):
    college_index = collegeData[collegeData['College Name'] == college].index[0]
    similar = similarity[college_index]

    # Get indices of top 10 most similar colleges (excluding the college itself)
    college_list = sorted(list(enumerate(similar)), reverse=True, key=lambda x: x[1])[1:11]

    # Print the recommended colleges
    for i in college_list:
        print(collegeData.iloc[i[0]]['College Name']+","+collegeData.iloc[i[0]]['City']+","+collegeData.iloc[i[0]]['State'])
```

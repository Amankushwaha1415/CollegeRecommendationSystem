```python
# here we applly stemming first
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# define a function to perform stemming on the tags column

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
stem(collegeData['tags2'][0])
```

```python
collegeData['tags3']=collegeData['tags2'].apply(stem)
```

```python
sorted(list(enumerate(similarity[78])),reverse=True, key=lambda x:x[1])[1:11]   #here we get top 6 sorted list of movies based on similarity
```

```python
# defining function for recommendation

def recommend(college):
    college_index = collegeData[collegeData['College Name'] == college].index[0]
    similar = similarity[college_index]

    # Get indices of top 10 most similar colleges (excluding the college itself)
    college_list = sorted(list(enumerate(similar)), reverse=True, key=lambda x: x[1])[1:11]

    # Print the recommended colleges
    for i in college_list:
        print(collegeData.iloc[i[0]]['College Name']+","+collegeData.iloc[i[0]]['City']+","+collegeData.iloc[i[0]]['State'])
```

```python
def unique_elements_from_column(df_column):
    unique_elements = set()
    for lst in df_column:
        unique_elements.update(lst)
    return unique_elements
```

```python
uniqueCourses = list(unique_elements_from_column(df_copy['Courses']))
```

```python
uniqueCourses
```

```python
len(uniqueCourses)
```

```python
# function to arrange the uniqueCourses list in a particular order where first Btech next comes BE then comes Mtech and then comes phd and then diploma then rest others 
def arrange_courses(courses):
    order = {
        'B.Tech': 1,
        'BE': 2,
        'M.Tech': 3,
        'ME': 4,
        'PhD': 5,
        'Diploma': 6,
        'B.Sc': 7,
        'M.Sc': 8
    }
    # Normalize capitalization (treat case-insensitive)
    normalized_courses = [c.strip() for c in courses]

    # Sort with duplicates preserved
    sorted_courses = sorted(
        normalized_courses,
        key=lambda x: (order.get(x, 1000), x)
    )
    return sorted_courses

arranged_courses = arrange_courses(uniqueCourses)
```
